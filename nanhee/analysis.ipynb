```python
import pandas as pd
import numpy as np
import os

data_path = './'

clinic_table = pd.read_csv(os.path.join(data_path, 'Clinical_Variables.csv')).to_numpy()[:,1:]
genetic_table = pd.read_csv(os.path.join(data_path, 'Genetic_alterations.csv')).to_numpy()[:,1:]
survival_table = pd.read_csv(os.path.join(data_path, 'Survival_time_event.csv')).to_numpy()[:,1:]
treatment_table = pd.read_csv(os.path.join(data_path, 'Treatment.csv')).to_numpy()[:,1:]

treatment = 1
no_treatmet = 0
treatment_index = np.where(treatment==treatment_table)[0] # 치료를 받은 사람들의 Index (총 488명)
no_treatment_index = np.where(no_treatmet==treatment_table)[0] # 치료를 받지 않은 사람들의 Index (총 512명)
print('Treatment :', treatment_index.shape[0])
print('No Treatment :', no_treatment_index.shape[0])
print('Total :', treatment_index.shape[0] + no_treatment_index.shape[0])
print()

survive = 0
no_survive = 1
survive_index = np.where(survive==survival_table[:,1])[0] # 생존한 사람들의 Index (총 109명)
no_survive_index = np.where(no_survive==survival_table[:,1])[0] # 생존하지 못 한 사람들의 Index (총 891명)
print('Survive :', survive_index.shape[0])
print('No Survive :', no_survive_index.shape[0])
print('Total :', survive_index.shape[0] + no_survive_index.shape[0])
print()

treatment_survive_index = np.intersect1d(treatment_index, survive_index) # 치료 O, 생존 O (총 48명)
treatment_no_survive_index = np.intersect1d(treatment_index, no_survive_index) # 치료 O, 생존 X (총 440명)
no_treatment_survive_index = np.intersect1d(no_treatment_index, survive_index) # 치료 X, 생존 O (총 61명)
no_treatment_no_survive_index = np.intersect1d(no_treatment_index, no_survive_index) # 치료 X, 생존 X (총 451명)
print('Treatment & Survive :', treatment_survive_index.shape[0])
print('Treatment & No Survive :', treatment_no_survive_index.shape[0])
print('No Treatment & Survive :', no_treatment_survive_index.shape[0])
print('No Treatment & No Survive :', no_treatment_no_survive_index.shape[0])
print('Total :', treatment_survive_index.shape[0] + treatment_no_survive_index.shape[0] + no_treatment_survive_index.shape[0] + no_treatment_no_survive_index.shape[0])
print()

print('########## Genetic Table ##########')
print()

treatment_survive_genetic = genetic_table[treatment_survive_index]
treatment_no_survive_genetic = genetic_table[treatment_no_survive_index]
no_treatment_survive_genetic = genetic_table[no_treatment_survive_index]
no_treatment_no_survive_genetic = genetic_table[no_treatment_no_survive_index]

print('Treatment & Survive')
sum_treatment_survive_genetic = np.sum(treatment_survive_genetic, axis=0)
max_genetic = np.max(sum_treatment_survive_genetic)
min_genetic = np.min(sum_treatment_survive_genetic)
print('Max : {} / Max Index : {}'.format(max_genetic, np.where(sum_treatment_survive_genetic == max_genetic)[0]))
print('Min : {} / Min Index : {}'.format(min_genetic, np.where(sum_treatment_survive_genetic == min_genetic)[0]))
print()

print('Treatment & No Survive')
sum_treatment_no_survive_genetic = np.sum(treatment_no_survive_genetic, axis=0)
max_genetic = np.max(sum_treatment_no_survive_genetic)
min_genetic = np.min(sum_treatment_no_survive_genetic)
print('Max : {} / Max Index : {}'.format(max_genetic, np.where(sum_treatment_no_survive_genetic == max_genetic)[0]))
print('Min : {} / Min Index : {}'.format(min_genetic, np.where(sum_treatment_no_survive_genetic == min_genetic)[0]))
print()

print('No Treatment & Survive')
sum_no_treatment_survive_genetic = np.sum(no_treatment_survive_genetic, axis=0)
max_genetic = np.max(sum_no_treatment_survive_genetic)
min_genetic = np.min(sum_no_treatment_survive_genetic)
print('Max : {} / Max Index : {}'.format(max_genetic, np.where(sum_no_treatment_survive_genetic == max_genetic)[0]))
print('Min : {} / Min Index : {}'.format(min_genetic, np.where(sum_no_treatment_survive_genetic == min_genetic)[0]))
print()

print('No Treatment & No Survive')
sum_no_treatment_no_survive_genetic = np.sum(no_treatment_no_survive_genetic, axis=0)
max_genetic = np.max(sum_no_treatment_no_survive_genetic)
min_genetic = np.min(sum_no_treatment_no_survive_genetic)
print('Max : {} / Max Index : {}'.format(max_genetic, np.where(sum_no_treatment_no_survive_genetic == max_genetic)[0]))
print('Min : {} / Min Index : {}'.format(min_genetic, np.where(sum_no_treatment_no_survive_genetic == min_genetic)[0]))
print()

print('###################################')

sum_treatment_survive_genetic_list = []
sum_treatment_no_survive_genetic_list = []
sum_no_treatment_survive_genetic_list = []
sum_no_treatment_no_survive_genetic_list = []
for i in range(1, 301):
    sum_treatment_survive_genetic_list.extend([i]*sum_treatment_survive_genetic[i-1])
    sum_treatment_no_survive_genetic_list.extend([i]*sum_treatment_no_survive_genetic[i-1])
    sum_no_treatment_survive_genetic_list.extend([i]*sum_no_treatment_survive_genetic[i-1])
    sum_no_treatment_no_survive_genetic_list.extend([i]*sum_no_treatment_no_survive_genetic[i-1])


import matplotlib.pyplot as plt
plt.subplot(221)
plt.hist(sum_treatment_survive_genetic_list, bins=300, label='Treatment & Survive')
plt.legend()
plt.subplot(222)
plt.hist(sum_treatment_no_survive_genetic_list, bins=300, label='Treatment & No Survive')
plt.legend()
plt.subplot(223)
plt.hist(sum_no_treatment_survive_genetic_list, bins=300, label='No Treatment & Survive')
plt.legend()
plt.subplot(224)
plt.hist(sum_no_treatment_no_survive_genetic_list, bins=300, label='No Treatment & No Survive')
plt.legend()
plt.show()
```

```python
print('########## Genetic Table 2##########')
print()

print('Treatment & Survive')
# print(sum_treatment_survive_genetic)

mean_genetic = np.mean(sum_treatment_survive_genetic)
var_genetic = np.var(sum_treatment_survive_genetic)
std_genetic = np.std(sum_treatment_survive_genetic)

print('Mean : {}'.format(mean_genetic))
print('Var : {}'.format(var_genetic))
print('Std : {}'.format(std_genetic))

print('High of Mean+Var Index : {}'.format(np.where(sum_treatment_survive_genetic > mean_genetic + var_genetic)[0]))
print('Low of Mean-Var Index : {}'.format(np.where(sum_treatment_survive_genetic < mean_genetic - var_genetic)[0]))

print('High of Mean+Std Index : {}'.format(np.where(sum_treatment_survive_genetic > mean_genetic + std_genetic)[0]))
print('Low of Mean-Std Index : {}'.format(np.where(sum_treatment_survive_genetic < mean_genetic - std_genetic)[0]))

print()
print()
print()


print('No Treatment & Survive')
# print(sum_treatment_no_survive_genetic)

mean_genetic = np.mean(sum_no_treatment_survive_genetic)
var_genetic = np.var(sum_no_treatment_survive_genetic)
std_genetic = np.std(sum_no_treatment_survive_genetic)

print('Mean : {}'.format(mean_genetic))
print('Var : {}'.format(var_genetic))
print('Std : {}'.format(std_genetic))

print('High of Mean+Var Index : {}'.format(np.where(sum_no_treatment_survive_genetic > mean_genetic + var_genetic)[0]))
print('Low of Mean-Var Index : {}'.format(np.where(sum_no_treatment_survive_genetic < mean_genetic - var_genetic)[0]))

print('High of Mean+Std Index : {}'.format(np.where(sum_no_treatment_survive_genetic > mean_genetic + std_genetic)[0]))
print('Low of Mean-Std Index : {}'.format(np.where(sum_no_treatment_survive_genetic < mean_genetic - std_genetic)[0]))

print()
print()
print()


print('Survive')
survive_genetic = genetic_table[survive_index]
sum_survive_genetic = np.sum(survive_genetic, axis=0)

mean_genetic = np.mean(sum_survive_genetic)
var_genetic = np.var(sum_survive_genetic)
std_genetic = np.std(sum_survive_genetic)

print('Mean : {}'.format(mean_genetic))
print('Var : {}'.format(var_genetic))
print('Std : {}'.format(std_genetic))

print('High of Mean+Var Index : {}'.format(np.where(sum_survive_genetic > mean_genetic + var_genetic)[0]))
print('Low of Mean-Var Index : {}'.format(np.where(sum_survive_genetic < mean_genetic - var_genetic)[0]))

print('High of Mean+Std Index : {}'.format(np.where(sum_survive_genetic > mean_genetic + std_genetic)[0]))
print('Low of Mean-Std Index : {}'.format(np.where(sum_survive_genetic < mean_genetic - std_genetic)[0]))

print()
print()
print()

'''
'Treatment & No Survive'의 결과는 분산이 크기때문에 /
중앙값처리로 분석해볼 필요가 있음

아래는 Test.

'''
medain_genetic = np.median(sum_no_treatment_survive_genetic)
var_genetic = np.var(sum_no_treatment_survive_genetic)
std_genetic = np.std(sum_no_treatment_survive_genetic)

print('Medain : {}'.format(medain_genetic))

print('High of Medain+Var Index : {}'.format(np.where(sum_no_treatment_survive_genetic >= medain_genetic + var_genetic)[0]))
print('Low of Medain-Var Index : {}'.format(np.where(sum_no_treatment_survive_genetic < medain_genetic - var_genetic)[0]))

print('High of Medain+Std Index : {}'.format(np.where(sum_no_treatment_survive_genetic >= medain_genetic + std_genetic)[0]))
print('Low of Medain-Std Index : {}'.format(np.where(sum_no_treatment_survive_genetic < medain_genetic - std_genetic)[0]))
```
